## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'

import pandas as pd

df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)

df_total = pd.merge(df1, df2, on='Customer', how='outer')
df_total = pd.merge(df_total, df3, on='Customer', how='outer')

df_total = df_total.drop(columns=['ST_x', 'GENDER_x', 'Education_x',
       'Customer Lifetime Value_x', 'Income_x', 'Monthly Premium Auto_x',
       'Number of Open Complaints_x', 'Policy Type_x', 'Vehicle Class_x',
      'Total Claim Amount_x', 'ST_y', 'GENDER_y', 'Education_y',
       'Customer Lifetime Value_y', 'Income_y', 'Monthly Premium Auto_y',
       'Number of Open Complaints_y', 'Total Claim Amount_y', 'Policy Type_y',
       'Vehicle Class_y'])

df_total.columns = df_total.columns.str.lower()

df_total['customer lifetime value'] = df_total['customer lifetime value'].apply(lambda x: round(x, 2))
df_total['total claim amount'] = df_total['total claim amount'].apply(lambda x: round(x, 2))
df_total['vehicle class'] = df_total['vehicle class'].apply(lambda x: 'Luxury' if x in ['Luxury SUV', 'Sports Car', 'Luxury Car'] else x)

df_total = df_total.dropna()

df_total['number of open complaints'] = df_total['number of open complaints'].astype(int)

df_total
```

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'

df = pd.read_csv(url)

#df.pivot(index = 'sales_channel', columns = 'monthly_premium_auto').stack()

channel_revenue = round(df.groupby('sales_channel')[['monthly_premium_auto']].sum(), 2)

channel_revenue

#in this case, we just needed to group all the data by different sale channels and sum all the values associated
#with those groups. We can see that the 'agent' channel is the most effective one, and web is the channel 
#that less revenue produces. 
```

```python
customer_value = df.pivot_table(index = 'gender', columns = 'education', values = ['customer_lifetime_value'], aggfunc = 'mean').stack()
customer_value['customer_lifetime_value'] = round(customer_value['customer_lifetime_value'], 2)

customer_value

#from the table below we can see that females with a level of education of high school diploma or below 
#are the group with higher customer lifetime value, where as for males, the group with highest customer
#lifetime value are the individuals with master's degree. 
```
